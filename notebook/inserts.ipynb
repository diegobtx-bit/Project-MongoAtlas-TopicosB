# Insertar Datos
- InserOne
- InsertMany

```python
#importando libreria
from connect import db

libros = db.libros

#Estructura coleccion (Equivalente Estructura Tabla)
libro = {
    'titulo': 'La casa Verde',
    'autor': 'Mario Vargas Llosa',
    'anio': 1965,
    'categorias': [
        'novela','ficcion'],
    'stock': 4
}

# InsertOne
res = libros.insert_one(libro)
print('InsertOne id: ', res.inserted_id)
```

```python
#InserMany (Varias inserciones)
docs = [
    #Libro 2
    {
    'titulo': 'Don Quijote',
    'autor': 'Miguel de Cervantes Saavedra',
    'anio': 1965,
    'categorias': [
        'clasico','ficcion'],
    'stock': 5
    },
    #Libro 3
    {
    'titulo': 'La ciudad y los perros',
    'autor': 'Mario Vargas Llosa',
    'anio': 1963,
    'categorias': [
        'novela'],
    'stock': 3
    }
]

res_many = libros.insert_many(docs)
print('InsertMany ids: ', res_many.inserted_ids)
```
